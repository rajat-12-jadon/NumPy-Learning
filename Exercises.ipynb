Valid Sudoku

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
s
```

```python
valid = True

# Check rows
for row in s:
    if np.sum(row) != 45 or len(np.unique(row)) != 9:
        valid = False
        break

# Check columns
if valid:
    for col in s.T:
        if np.sum(col) != 45 or len(np.unique(col)) != 9:
            valid = False
            break

# Check 3x3 boxes
if valid:
    for i in range(0, 9, 3):
        for j in range(0, 9, 3):
            box = s[i:i+3, j:j+3].flatten()

            if np.sum(box) != 45 or len(np.unique(box)) != 9:
                valid = False
                break

if valid:
    print("Valid Sudoku")
else:
    print("Invalid Sudoku")
```
