```python
# NumPy Indexing and Slicing of Matrix
```

```python
arr = np.arange(1,31).reshape(5, 6)
arr
```

```python
arr[0]
```

```python
arr[0,0]
```

```python
arr[0,4]
```

```python
arr[0][1]
```

```python
slice = arr[0:2, 1:3]
slice
```

```python
arr[2:6, 4:6]
```

```python
arr[:,2]
```

```python
# Boolean Indexing
```

```python
arr = np.arange(11, 21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
