```python
import numpy as np
```

```python
a1 = np.array([1, 2, 3, 4, 5])
a2 = np.array([6, 7, 8, 9, 10])
```

```python
sum = a1 + a2
sum
```

```python
diff = a2 - a1
diff
```

```python
product = a1 * a2
product
```

```python
a1 / a2
```

```python
a1 // a2
```

```python
a1 ** a2
```

```python
# Deep and Shallow Copy
```

```python
a = np.arange(1, 21)
a
```

```python
slice = a[: 5]
slice * 10
slice
```

```python
a
```

```python
a[4 : 7] * 10
a
```

```python
a
b = a
b[0] = 99
a
```

```python
b = a.copy()
b
```

```python
b[0] = 1
b
```
